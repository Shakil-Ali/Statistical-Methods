> 2. (a) Explain why the height of the density curve is 1/(b ‚àí a). (Hint: Re-
member that the total area under the density curve must equal 1,
and recall the formula for the area of a rectangle.)

The total area under the curve must be 1. The base has length $b-a$. The area of a rectangle is base $\times$ height. And so if the height is $h$

$$
\begin{aligned}
1 & = (b-a)h \\
\Rightarrow
h & = \frac{1}{b-a}
\end{aligned}
$$

> 2. (b) What is P {X ‚â§ (a + b)/2}?

The value $\frac{a+b}{2}$ lies exactly halfway along the rectangles length. As it is a symmetric distribution, half the area under the probability density function lies below this value and:

$$\Pr\left(X \leq  \frac{a+b}{2}\right) = 1$$

> 3. (a) P {X > 1/3}

The probability distribution is a rectangle whose height is $h = \frac{1}{1} = 1$. The probability

$$
\Pr\left(X > \frac{1}{3}\right) 
= \Pr\left(\frac{1}{3} < X < 1 \right) 
$$

As everywhere other than the interval $(0,1)$ the probability density is zero. This is a rectangle whose base is $1 - \frac{1}{3} = \frac{2}{3}$ and whose height is $1$. So the area is:

$$
\Pr\left(\frac{1}{3} < X < 1 \right) 
= \frac{2}{3} \cdot 1 = \frac{2}{3}
$$

> 3. (b) P {X ‚â§ 0.7}

Following a similar argument:
$$
\begin{aligned}
\Pr\left(X \geq 0.7 \right)
& = \Pr\left(0.7  \leq X < 1 \right) \\
& = \Pr\left(0.7  < X < 1 \right) \\
& = (1- 0.7) \cdot 1 = 0.3
\end{aligned}
$$


> 3. (c) P {0.3 < X ‚â§ 0.9}

Following a similar argument:
$$
\begin{aligned}
\Pr\left(0.3  < X \leq 0.9 \right)
& = (0.9- 0.3) \cdot 1 = 0.6
\end{aligned}
$$

> 3. (d) P {0.2 ‚â§ X < 0.8}

Following a similar argument:
$$
\begin{aligned}
\Pr\left(0.2  \leq X < 0.8 \right)
& = (0.8- 0.2) \cdot 1 = 0.6
\end{aligned}
$$

> 8. (a) What is the height of the curve at the value 2? (Hint: You will have to recall the formula for the area of a triangle.)

The formula for the area of a triangle of base $b$ and height $h$ is:

$$\text{area } = \frac{b h}{2}$$

For us $b = 4-0 = 4$ and the area must be $1$, so 

$$h = \frac{2}{b}\cdot 1 = \frac{1}{2}$$

This is the height of the curve above $2$ and so we can say that if our probability density function is $p_Y$ then:

$$p_Y(2) = \frac{1}{2}$$

> 8. (b) What is the probability she will study more than 3 hours?

This is the area under $p_Y$ between $3$ and $4$. It is a triangle whose base is $4-3 = 1$ and whose height is $\frac{h}{2} = \frac{1}{4}$ (it is halfway between the apex, at height $h$,  and the corner, at height $0$).
The area of this triangle is $\frac{1}{2}\cdot 1 \cdot \frac{1}{4} = \frac{1}{8}$. And so:

$$\Pr(Y > 3) = \frac{1}{8}$$

> 8. (c) What is the probability she will study between 1 and 3 hours?

There are a number of ways of calculating this. One way to do so is to recognise that from symmetry:

$$\Pr(Y < 1)  = \Pr(Y > 3) = \frac{1}{8}$$

And by mutual exclusivity of events:

$$\Pr( 1 < Y < 3)  = 1 - \Pr(Y < 1)- \Pr(Y > 3) = 1 - \frac{1}{8} - \frac{1}{8} = \frac{3}{4}$$



## Using `plot` for functions

The `plot` function from `matplotlib.pyplot` takes a sequence of x-coordinates and a sequence of y-coordinates and draws straight lines between the corresponding sequence of (ùë•,ùë¶) points. This can be used to plot frequency polygons (as seen in week 2) but can also be used to draw known functions.

For our purposes, a known function $f:(a,b)\to \mathbb{R}$ is one where you know the value of the function for all inputs over some interval of the real line, $(a,b) \subseteq \mathbb{R}$. We will differentiate between piecewise linear functions (those made up of a finite number of straight lines) and smooth functions (which may contain curved lines).

`plot` can be used to plot both piecewise linear and smooth functions. For smooth functions, it essentially tricks the eye, by drawing lots of very short linear sections that approximate the smooth function. Look at the two examples in the code block below and try to anticipate the shape of the resulting plot. Experiment with the arguments to `np.linspace` to see what effect it has on the second plot.

Note that the second uses `numpy` array's elementwise arithmetic. Remind yourself of that by looking at the link [here](https://scipy-lectures.org/intro/numpy/operations.html#elementwise-operations) or returning to your notebook for week 3.

```python
## a piecewise linear function plot
# the sequence of x coordinates as a numpy array
xs = np.array([0, 1, 2, 3, 4, 5, 6])
# the sequence of y coordinates as a numpy array
ys = np.array([0, 1, 1, 2, 1, 1, 0])
plt.figure()
plt.plot(xs, ys, '-', lw=2)
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
# ensure the base of the plot is at y=0
plt.ylim(0, None)

## a "smooth" function plot
# to trick the eye we need to evalute our function at lots
# of closely packed points. Let's first create the x=values
# at which we want to evalute the function
lower_bound = -5
upper_bound = 5
number_of_points = 101
xs = np.linspace(lower_bound, upper_bound, number_of_points)
# the sequence of y coordinates are calculated at every x-value
# one of the most effective ways to do this is to write a function
# make sure that your function works with inputs that are numpy arrays
def f(xs):
    return xs**3 - 4*xs**2 + xs + 6
    
ys = f(xs)
plt.figure()
plt.plot(xs, ys, '-', lw=2)
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```

```python
## 4. The scores on a scholastic achievement test were normally distributed
# with mean 520 and standard deviation 94.
# If your score is RV X then its mean and standard deviation are:
mu = 520.
sigma = 94.

# (a) If your score was 700, by how many standard deviations did it exceed the average score?
# number of standard deviations above the mean is:
num_sds = (700.-mu)/sigma
print(f"num_sds = {num_sds:.2f}")

# (b) What percentage of examination takers received a higher score than you did?
# From Table:
# Pr(X > 700 ) = Pr(Z > (700-520)/94)) = Pr(Z > 1.91) = 1  - Pr(Z > 1.91) = 1 - 0.9719
table_p = 1 - 0.9719
print("4. (b) from table:") 
print(f"Pr(X > 700 ) = Pr(Z > {num_sds:.2f}) = {table_p:.4f}")
# From scipy.stats.norm.cdf
calc_p = 1 - stats.norm.cdf(num_sds)
print("4. (b) from scipy:") 
print(f"Pr(X > 700 ) = Pr(Z > {num_sds:.4f}) = {calc_p:.4f}")
print("Note that the differences are due to the increased precision of sciypy.")
print("The table only allows us to specify z values to 2 decimal places.")
print()

## 6. The life of a certain automobile tire is normally distributed with mean
# 35,000 miles and standard deviation 5000 miles.
# Model as a normal RV X with mean and sd:
mu = 35000
sigma = 5000

# (a) What proportion of such tires last between 30,000 and 40,000 miles?
a = 30000
b = 40000
za = (a-mu)/sigma
zb = (b-mu)/sigma
# Pr(30000 < X < 40000) = Pr(X < 40000) - Pr(X < 30000)
# So we need these two probabilities
print("6. (a)")
print(f"Pr(X < 30000) = Pr(Z < {za:.2f}) = 1  - Pr(Z < {-za:.2f})")
print(f"Pr(X < 40000) = Pr(Z < {zb:.2f})")
# From the table these values are:
table_pa = 1 - 0.8413
table_pb = 0.8413
# and
print("6. (a) from table:") 
print(f"Pr(30000 < X < 40000) = Pr(X < 40000) - Pr(X < 30000) = {table_pb:.4f} - {table_pa:.4f} = {table_pb-table_pa:.4f}")
# From scipy:
calc_pa = 1 - stats.norm.cdf(-za)
calc_pb = stats.norm.cdf(zb)
print("6. (a) from scipy:") 
print(f"Pr(X < 30000) = Pr(Z < {za:.2f}) = 1  - Pr(Z < {-za:.2f}) = {calc_pa:.4f}")
print(f"Pr(X < 40000) = Pr(Z < {zb:.2f}) = {calc_pb}")
print(f"Pr(30000 < X < 40000) = Pr(X < 40000) - Pr(X < 30000) = {calc_pb-calc_pa:.4f}")
print()
# They agree, but this isn't quite the 0.68 we expected from the approximation rule.

# (b) What proportion of such tires last over 40,000 miles?
print(f"Pr(X > 40000) = Pr(Z > {zb:.2f}) = 1 - Pr(Z < {zb:.2f}) = {1 - calc_pb:.4f}")
# we calculated this already.
print()

# (c) What proportion last over 50,000 miles?
c = 50000
zc = (c-mu)/sigma
table_pc = 1- 0.9987
calc_pc = 1 - stats.norm.cdf(zc)
print(f"Table: Pr(X > {c}) = Pr(Z > {zc:.2f}) = 1 - Pr(Z < {zc:.2f}) = {table_pc:.4f}")
print(f"Scipy: Pr(X > {c}) = Pr(Z > {zc:.2f}) = 1 - Pr(Z < {zc:.2f}) = {calc_pc:.4f}")
# we calculated this already.
print()

## 7. Suppose you purchased such a tire as described in Prob. 6. If the tire is in
# working condition after 40,000 miles, what is the conditional probability
# that it will still be working after an additional 10,000 miles?
# This is conditional probability: Pr(X > 50000 | X > 40000)
# Using the Product rule: 
print("Pr(X > 50000 | X > 40000) = Pr(X > 50000 and X > 40000)/Pr(X > 40000) = Pr(X > 50000)/Pr(X > 40000)")
# Because the combined event (X > 50000 and X > 40000)
# Is identical to event (X > 50000) and so
print(f"Pr(X > 50000 | X > 40000) = Pr(Z > {zc})/Pr(Z > {zb}) = {calc_pc:.4f}/{calc_pb:.4f} = {calc_pc/calc_pb:.4f}")
print()

## Addendum: We can also calcualte this directly in Sciyp without converting to Z
print("Addendum:")
# 4. (b)
print("4. (b) from scipy no conversion:") 
mu = 520
sigma = 94
# here we calculate it on the fly
print(f"Pr(X > 700) = {1 - stats.norm.cdf(700, loc=mu, scale=sigma):.4f}")
# here we create a frozen distribution first:
rv = stats.norm(loc=mu, scale=sigma)
print(f"Pr(X > 700) = {1 - rv.cdf(700):.4f} -- with frozen distribution")
print()

# 6.
print("6. (a) from scipy no conversion:") 
mu = 35000
sigma = 5000
a = 30000
b = 40000
# here we calculate it on the fly
print(f"Pr(30000 < X < 40000) = {stats.norm.cdf(b, loc=mu, scale=sigma)-stats.norm.cdf(a, loc=mu, scale=sigma):.4f}")
# here we create a frozen distribution first:
rv = stats.norm(loc=mu, scale=sigma)
print(f"Pr(30000 < X < 40000) = {rv.cdf(b)-rv.cdf(a):.4f} -- with frozen distribution")
print()
```
